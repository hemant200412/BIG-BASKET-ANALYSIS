```python
# Sort top & least products
top_products = data.sort_values(by="rating", ascending=False).head(10)
least_products = data.sort_values(by="rating", ascending=True).head(10)

# Display nicely as DataFrames
from IPython.display import display

print("✅ Top 10 Products by Rating")
display(top_products[['product', 'brand', 'rating']].reset_index(drop=True))

print("\n❌ Least 10 Products by Rating")
display(least_products[['product', 'brand', 'rating']].reset_index(drop=True))
```

```python
# Create a new column for discount percentage
data['discount_pct'] = ((data['market_price'] - data['sale_price']) / data['market_price']) * 100

# Show top 10 products with the highest discount
top_discounts = data.sort_values(by="discount_pct", ascending=False).head(10)

print("Top 10 Products with Highest Discount:")
display(top_discounts[['product', 'brand', 'market_price', 'sale_price', 'discount_pct']].reset_index(drop=True))
```

```python
# Count missing values in each column
missing_values = data.isnull().sum()

# Show only columns with missing values
missing_values = missing_values[missing_values > 0]

print("Missing Values in Each Column:")
print(missing_values)
```

```python
# Select only numeric columns
numeric_cols = data.select_dtypes(include=[np.number]).columns

print("🔍 Detecting Outliers using IQR Method...\n")

outliers_summary = {}

# Step 1: Detect Outliers
for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)][col]
    
    outliers_summary[col] = len(outliers)

# Show Outlier Counts
print("📌 Outliers detected in each numeric column:")
for col, count in outliers_summary.items():
    print(f"{col}: {count} outliers")
```

```python
#Replace Outliers with Mean
for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    mean_val = data[col].mean()
    
    data.loc[(data[col] < lower_bound) | (data[col] > upper_bound), col] = mean_val
```
